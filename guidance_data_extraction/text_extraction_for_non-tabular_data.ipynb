# Text Extractor for Non-tabular Data

```python
import re
from bs4 import BeautifulSoup, NavigableString
```

```python
guidance_dir = '../guidance_page_collection/guidance_pages_annotated/'
NUM = 59
```

```python
output_dir_path = 'extracted_texts/'
```

```python
ignored_tags = [ 'table', 'footer', 'script' ]

unwrapped_tags = [ 'span', 'a', 'strong', 'b', 'cite', 'em', 'i', 'mark', 'small', 'sub', 'sup' ]
```

```python
def analyze(guidance):
    html_content = open(f'{guidance_dir}{guidance}.html', 'r').read()
    html_content = re.sub('\n+', ' ', html_content)
    html_content = re.sub('\s+', ' ', html_content)
    soup = BeautifulSoup(html_content, 'html.parser')

    body = soup.find_all('body')
    print(f'{len(body) = }')
    assert len(body) == 1, f'{len(body) = } should be 1'

    body = body[0]

    for ignored in body.find_all(ignored_tags):
        ignored.decompose()

    for hidden in body.find_all(style=re.compile(r'display:\s*none')):
        hidden.decompose()

    for ul in body.find_all('ul'):
        ul_text = ''
        for li in ul.find_all('li'):
            li_text = li.text.rstrip(' ')
            li_text = re.sub('\s+', ' ', li_text)
            if (li_text.endswith('.') or li_text.endswith('?')):
                ul_text += li_text + ' '
            elif (len(li_text) > 0):
                ul_text += li_text + ', '
            li.decompose()
        ul.string = ul_text.rstrip(' ').rstrip(',').rstrip('.')+'.\n'

    for tag in body.find_all(unwrapped_tags):
        tag.unwrap()

    body = BeautifulSoup(str(body), 'html.parser')

    results = ''
    for line in body.get_text('\n').split('\n'):
        texts = line.strip()
        if (texts.find(' ') < 0):
            continue

        if (texts.endswith(':')):
            results += texts + ' '
        else:
            results += texts+'\n'

    with open(f'{output_dir_path}{guidance}_non-tab.txt', 'w') as f:
        f.write(results)
```

```python
for target in range(1, NUM + 1):
    print(f'{target = }')
    analyze(target)

print('Done')
```
